### Memorial de cálculo
Sabe-se que:

\begin{equation*}
\rho  = C \cdot N \cdot e  \\
e = diametro do fio = \frac{alturaSuporte}{N} \\
N = número de voltas da bobina  \\
\rho = comprimento do fio \\
e = diametro do fio = \frac{alturaSuporte}{N} \\
C = perimetro  \\
\end{equation*}

```python
import math
```

```python
# ------- Alguns parâmetros -------
v = 12 # Tensão em [V]
mi = [1.5e4, 1e4, 5e3] # Possíveis valores de ferrite, foi utilizado os valores máximos de cada

# ------- Suporte da bobina -------
comp_sup = (102.042)# Comprimento em mm
larg_sup = 12.048# largura em mm
per = ((larg_sup * 2) + (comp_sup * 2))/1000# perimetro em m
alt_sup = 14.49/1000 #altura em m

# ------- Fio 15 AWG-------
d_fio = 1.450/1000 #Diâmetro em metro
comp_fio = per * alt_sup # comprimento fio
resis_fio =10.3/1000 #ohm/metro

# ------- Bobina -------
numero_voltas = alt_sup/d_fio # Numero de voltas

# ------- Outros parametros -------
R = resis_fio * numero_voltas * per
```

```python
print("Perimetro:",per,"[m]")
print("Para fio 15 AWG:")
print("Número de voltas:",int(numero_voltas))
print("Comprimento do fio:",(int(numero_voltas))*alt_sup,"[m]")
print("Resistência:",R,"[Ω]")
print("Corrente: ",v/R,"[A]")
```

```python
# ------- Fio 35 AWG -------
d_fio2= 0.14/1000 #Diâmetro em metro
comp_fio2 = per * alt_sup # comprimento fio
resis_fio2 = 1.1 #ohm/metro

# ------- Bobina -------
numero_voltas2 = alt_sup/d_fio2 # Numero de voltas

# ------- Outros parametros -------
R2 = resis_fio2 * numero_voltas2 * per
```

```python
print("Para fio 35 AWG:")
print("Número de voltas:",int(numero_voltas2))
print("Comprimento do fio:",(int(numero_voltas2))*alt_sup,"[m]")
print("Resistência:",R2,"[Ω]")
print("Corrente: ",v/R2,"[A]")
```
